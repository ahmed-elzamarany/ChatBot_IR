First we create model to sentimate analysis

```python
import nltk
import string
#Importing the stopwords
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import re
import pickle
```

```python
# tokenized_dataset=pd.concat([sad_df,surprised_df]).reset_index()
tokenized_dataset=Alldataset.copy()
tokenized_dataset=tokenized_dataset[['context', 'utterance']]
tokenized_dataset.utterance=tokenized_dataset.utterance.apply(lambda row: row.replace("_comma_",""))
# tokenized_dataset['tokenized_text'] = tokenized_dataset.utterance.apply(lambda row: nltk.word_tokenize(row.lower()))
tokenized_dataset
```

```python
#Initializing the WordNetLemmatizer
lemmer = nltk.stem.WordNetLemmatizer()
#A dictionary reference for replacing special characters
remove_punct_dict = dict((ord(punct), None) for punct in string.punctuation)
def LemTokens(tokens):
       return [lemmer.lemmatize(token,'v') for token in tokens if token not in set(stopwords.words('english'))]
def LemNormalize(text):
       return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punct_dict)))
tokenized_dataset.drop_duplicates(subset ="utterance",keep = "first", inplace = True)
tokenized_dataset['tokenized_text']=tokenized_dataset.utterance.apply(lambda row: LemNormalize(row))
```

```python
utterance=tokenized_dataset['utterance'].values
# tokenized=tokenized_dataset['tokenized_text'].values.tolist()
# dic=dict(zip(utterance, zip(tokenized)))
```

```python
vectorizer = CountVectorizer(tokenizer = LemNormalize, ngram_range=(2,2))
```

```python
X = vectorizer.fit_transform(np.array(utterance))
features_count=np.sum(np.array(X.toarray()), axis=0)
```

```python
data=np.vstack((np.array(vectorizer.get_feature_names()),features_count.T))
```

```python
df_features_count = pd.DataFrame(data=data.T, columns=["feature","count"])
df_features_count
```

```python
def hi(text):
    if(re.search(r'\d', text)):
        return np.nan
    else:
        return text

df_features_count["feature"]=df_features_count["feature"].apply(lambda row: hi(row))
df_features_count= df_features_count.dropna()
df_features_count=df_features_count.reset_index()
df_features_count=df_features_count[['feature', 'count']]
```

```python
df_features_count
```

```python
indeces_to_delete=np.where(np.isin(np.array(vectorizer.get_feature_names()),df_features_count['feature'].values)==False)
indeces_to_delete
```

```python
Y = np.delete(X.toarray(), indeces_to_delete, axis=1)
print(Y.shape)
# print(X.shape)
```

```python
classifier = LogisticRegression()
y_labels=tokenized_dataset['context'].values.tolist()
classifier.fit(Y, y_labels)
```

```python
accuracy = classifier.score(Y, y_labels)
print("“Accuracy:", accuracy)
```
